```python
# pesquisa binária (em vetores ordenados)

def busca_binaria(lista, alvo):
    baixo = 0
    alto = len(lista) - 1

    while baixo <= alto:
        meio = (baixo + alto) // 2
        
        if lista[meio] == alvo:
            return meio  # Elemento encontrado
        elif lista[meio] < alvo:
            baixo = meio + 1  # Busca na metade direita
        else:
            alto = meio - 1  # Busca na metade esquerda
            
    return -1  # Elemento não encontrado

# Exemplo de uso:
minha_lista = [1, 3, 5, 7, 9, 11]
alvo = 7
resultado = busca_binaria(minha_lista, alvo)
print(f"Índice do elemento: {resultado}") # Saída: 3
```

```python
import numpy as np

class VetorNaoOrdenado :
    def __init__ ( self , capacidade ):
        self . capacidade = capacidade
        self . ultima_posicao = -1
        # array vazio de inteiros
        self . valores = np . empty ( self . capacidade , dtype = int )

# imprimir os valores - O ( n )
    def imprime ( self ) :
        if self . ultima_posicao == -1:
            print( " Vetor vazio ! " )
        else:
            for i in range ( self . ultima_posicao + 1):
                print(f' { i } - { self . valores [ i ]}')
```
